```python
import pandas as pd
```

```python
df = pd.DataFrame(pd.read_excel("data_1.xlsx"))
first_data_index = 500 #여기 수정
last_data_index= 650 # 여기 수정
insert_df= df["Peptide"][first_data_index: last_data_index]
insert_df
```

```python
print(f"output_{first_data_index}_{last_data_index}.xlsx")
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys 
from selenium.webdriver.support.ui import Select
import time

peptide_result_array = []
for index, value in enumerate(insert_df):
    peptide_dict = {}
    peptide_dict["Peptide"] = value.replace(u'\xa0', u'')
    options = webdriver.ChromeOptions()
    options.add_argument("headless")
    driver = webdriver.Chrome("./chromedriver", options=options)
    driver.get("https://flybase.org/blast/")


    elem = driver.find_element_by_id("sequence")
    time.sleep(1)
    print(value.replace(u'\xa0', u'') +"  검사를 시작합니다.")
    elem.send_keys(value.replace(u'\xa0', u''))
    time.sleep(1)
    select = Select(driver.find_element_by_id('database'))
    time.sleep(1)
    select.select_by_value("org/translation")
    time.sleep(1)

    driver.find_element_by_xpath("//input[@value='BLAST']").click() 
    time.sleep(30)
    def hasxpath(xpath):
        try: 
            driver.find_element_by_xpath(xpath)
            return True
        except:
            return False
    

    is_xpath = hasxpath("//div[@class='bold']/a")
    if (is_xpath):
        link = driver.find_element_by_xpath("//div[@class='bold']/a")
        link.click()
        time.sleep(10)
        the_value = driver.find_element_by_css_selector("body > div:nth-child(3) > div > div.col-sm-10.col-xs-12 > div > div:nth-child(6) > div > div:nth-child(2)");
        time.sleep(1)
        peptide_dict["Result"] = the_value.text
    else: 
        time.sleep(1)
        peptide_dict["Result"] = "N/A"
        
    peptide_dict
    
    peptide_result_array.append(peptide_dict)
    
    print( str(index +1) +"번 돌았음")
    driver.quit()
    
print(peptide_result_array)
data = pd.DataFrame(peptide_result_array)
data.to_excel(f"output_{first_data_index}_{last_data_index}.xlsx",encoding='utf8', index=False)
```
